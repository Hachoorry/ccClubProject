```python
# Colab 進行matplotlib繪圖時顯示繁體中文
# 下載台北思源黑體並命名taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

import pandas as pd # 額外加的
import matplotlib as mpl
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager

# 改style要在改font之前
# plt.style.use('seaborn')  

fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
mpl.rc('font', family='Taipei Sans TC Beta')
```

```python
#　存取雲端硬碟
from google.colab import drive
drive.mount('/content/drive')

# 引入資料
df = pd.read_csv(f'/content/drive/MyDrive/ccClubProject/csv/100_110年度串接縣市鄉鎮代號.csv')  
df.insert(0,'稻穀總產量(萬噸)',df['稻穀總產量']/100000000)  # 想避免科學記號
df.head(20)
```

```python
# 完成
table02 = pd.pivot_table(df, index=['年份'],columns = ['縣市'], values='稻穀總產量(萬噸)', aggfunc = 'sum')  # 各縣市各年度表現(所有作物加總)
table02.plot(kind='line', figsize=(19,9)) 
plt.legend()  # 標籤
plt.ylabel('稻穀總產量(萬噸)')
plt.title('各縣市各年度表現')  #  圖表名

plt.show()  # 出現圖，註解/不註解有不同顯示方式
```

```python
# 12/26更新完成
table03 = pd.pivot_table(df, index=['年份'],columns = ['作物'], values='稻穀總產量(萬噸)', aggfunc = 'sum')  # 處理資料(樞紐分析)
# table03

table03.plot(kind='line', figsize=(12,4),linewidth = 1.5 ) # 12/26加粗

plt.title('四總作物年趨勢')  #  圖表名
plt.ylabel('稻穀總產量(萬噸)')
```
